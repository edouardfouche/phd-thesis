```python
main_data = pandas.read_csv(file, header=0,index_col=0)
main_data.index = pandas.DatetimeIndex(main_data.index)
```

```python
main_data
```

```python
attr1 = "CL 32305 XQ01"
attr2 = "CT 32219 XQ01"
```

```python
main_data[attr1 + " N"] = (main_data[attr1] - main_data[attr1].min())/(main_data[attr1].max() - main_data[attr1].min())
main_data[attr2 + " N"] = (main_data[attr2] - main_data[attr2].min())/(main_data[attr2].max() - main_data[attr2].min())
```

```python
plt.figure()
attr = attr1 + " N"
main_data[attr].index = pandas.to_datetime(main_data[attr].index)
main_data[attr].plot(kind="line", title="Filling Level of the Flue Gas Cyclone (%)", figsize=(12,4))
plt.tight_layout()
```
