```python
import math

import matplotlib as mpl
mpl.rcParams['text.usetex'] = True 
mpl.rcParams['text.latex.preamble'] = r'\usepackage{libertine}' 
mpl.rc('font', family='serif')

import numpy as np
import pandas as pd
import seaborn as sns
from scipy.spatial import distance_matrix

import matplotlib.pyplot as plt
from matplotlib import rc

from cycler import cycler


mpl.rcParams['ps.usedistiller'] = 'xpdf' 

plt.style.use('seaborn-notebook')
# sns.set_palette("Blues_r")

plt.rcParams['axes.titlesize'] = '25'
plt.rcParams['axes.labelsize'] = '25'
plt.rcParams['legend.fontsize'] = '15'
plt.rcParams['xtick.labelsize'] = '20'
plt.rcParams['ytick.labelsize'] = '20'

monochrome=(cycler('color', sns.color_palette("husl", 8))*2+(cycler('marker', ['v', 's', "o"])*7)[0:16])
plt.rc('axes', prop_cycle=monochrome)

pd.options.display.max_rows = 999
```

```python
y1 = [0 if x < 10 else 1 for x in range(20)]
```

```python
plt.figure(figsize=(5,5))
plt.plot(range(len(y1)), y1)
plt.title("Sudden")
plt.axis('off')
```

```python
from math import exp
from numpy import random
def sig(x): return 1/(1+exp((1/5)*-x))
y2 = [random.choice([0,1], p=[1-sig(x-25),sig(x-25)]) for x in range(50)]
```

```python
plt.figure(figsize=(5,5))
plt.plot(range(len(y2)), y2, color="red")
plt.title("Gradual")
plt.axis('off')
```

```python
y3 = [(int(x/5)*10) for x in range(50)]
```

```python
plt.figure(figsize=(5,5))
plt.plot(range(len(y3)), y3,  color="green")
plt.title("Incremental")
plt.axis('off')
```

```python
s = [x%3 for x in range(10)]
y4 = sum([[x] * 5 for x in s], [])
```

```python
plt.figure(figsize=(5,5))
plt.plot(range(len(y4)), y4, color="orange")
plt.title("Recurrent")
plt.axis('off')
```

```python
uiucblue = (19/255,42/255,76/255)
uiucred = (232/255,74/255,39/255)
kitgreen = (50/255,161/255,137/255)
```

```python
import matplotlib.gridspec as gridspec

# Create 2x2 sub plots
gs = gridspec.GridSpec(1, 4, hspace=0.1)
plt.figure(figsize=(12,3))

ax = plt.subplot(gs[0, 0]) # row 0, col 0
plt.plot(range(len(y1)), y1, color=uiucblue, marker='')
plt.title("Sudden")
plt.axis('off')

ax = plt.subplot(gs[0, 1]) # row 0, col 1
plt.plot(range(len(y2)), y2, color=uiucred, marker='')
plt.title("Gradual")
plt.axis('off')

ax = plt.subplot(gs[0, 2]) # row 1, col 0
plt.plot(range(len(y3)), y3, color=kitgreen, marker='')
plt.title("Incremental")
plt.axis('off')

ax = plt.subplot(gs[0, 3]) # row 1, col 1
plt.plot(range(len(y4)), y4, color="black", marker='')
plt.title("Recurrent")
plt.axis('off')

#plt.tight_layout()
plt.subplots_adjust(top=0.8)

plt.savefig("conceptdrift_example_row.pdf")
```
